```python
%%writefile game_info.py

# Variable (Constant)
MAX_HP = 100

# Function
def announce_winner(player_name):
    return f"🏆 The winner is {player_name}!"

# Class
class Player:
    def __init__(self, name, hp):
        self.name = name
        self.hp = hp

    def take_damage(self, amount):
        self.hp -= amount
        return f"{self.name} now has {self.hp} HP"
```

```python
import game_info as gi

gi.announce_winner("Arthur")
```
