**과제 1**

Faker를 사용하여 1000명의 데이터를 class를 이용하여 만들어라



```python
pip install faker
```

```python
from faker import Faker
fake = Faker()
```

```python
class User:
    def __init__(self):
        self.name = fake.name()
        self.email = fake.email()
        self.job = fake.job()

    def __repr__(self):
        return f"<{self.name}: {self.job}>"
```

```python
users = [User() for _ in range(1000)]

print(f"Successfully generated {len(users)} users.")
```

```python
# 가독성을 고려하여 상위 30명의 샘플 데이터만 출력합니다.
for user in users[:30]:
    print(user)
```

**과제 2**

Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)

```python
class Chaser:
    def __init__(self, name):
        self.name = name
        self.target = None
        self.follow_count = 0
        self.status = "Idle"

    def following_mode(self, target_name):
        self.target = target_name
        self.status = "Following"
        self.follow_count += 1

        print(f"[{self.follow_count}] {self.name} is now in {self.status} mode, chasing {self.target}!")
```

```python
tom = Chaser("Tom")

print(f"--- Action Start: {tom.name} is looking for Jerry ---")

for i in range(1, 11):
    tom.following_mode("Jerry")
```

```python
print("-" * 30)
print(f"Object Name: {tom.name}")
print(f"Final Target: {tom.target}")
print(f"Total Follow Count: {tom.follow_count}")
print(f"Final Status: {tom.status}")
```
