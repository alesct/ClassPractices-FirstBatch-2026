```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "http://books.toscrape.com/catalogue/page-1.html"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


titles = []
items = soup.find_all('h3')

for item in items:
    link = item.find('a')
    if link and link.has_attr('title'):
        titles.append(link['title'])

df = pd.DataFrame(titles, columns=['Book Titles'])


print(df.head(20))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "http://books.toscrape.com/catalogue/page-1.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


titles = []
net_prices = []


products = soup.find_all('article', class_='product_pod')

for book in products:


    current_title = book.h3.a['title']
    titles.append(current_title)

    current_net_price = book.find('p', class_='price_color').text
    net_prices.append(current_net_price)


# CLEANING PROCESS
clean_prices = []
for p in net_prices:

    cleaned_p = re.sub(r'[^\d.]', '', p)
    clean_prices.append(float(cleaned_p))

df = pd.DataFrame({
    'Titles': titles,
    'Price (£)': clean_prices
})
print("---Price List ---")
print(df)


print("\n--- Price Summary ---")
print(df['Price (£)'].describe())
```
